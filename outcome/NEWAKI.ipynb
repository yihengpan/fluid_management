```python
import pandas as pd
chlo = "chloride.csv"
sofa = "sofa_pan.csv"
aki = "aki_final.csv"

sofa = pd.read_csv(sofa)
chl = pd.read_csv(chlo)
aki = pd.read_csv(aki)
```

```python
aki1 = aki[['icustay_id', 'day', 'aki_stage']]
aki1.head()
```

```python
icuid = []
for icu in aki1['icustay_id']:
    icuid.append(icu)
```

```python
my_Dict = {}
def new(icu_id):

    x = aki1[aki1['icustay_id'] == icu_id]

    if x[x['day'] == 1]['aki_stage'].any() == 0 and x[x['day'] >= 3]['aki_stage'].any() > 0:
        my_Dict[icu_id] = 1
    elif x[x['day'] == 1]['aki_stage'].any() == 1 and x[x['day'] >= 3]['aki_stage'].any() > 1:
        my_Dict[icu_id] = 1
    elif x[x['day'] == 1]['aki_stage'].any() == 2 and x[x['day'] >= 3]['aki_stage'].any() > 2:
        my_Dict[icu_id] = 1
    elif len(x[x['aki_stage'] >=3]) >= 3:
        my_Dict[icu_id] = 1
    else:
        my_Dict[icu_id] = 0
```

```python
for icu_id in icuid:
    new(icu_id)
```

```python
df = pd.DataFrame(list(my_Dict.items()), columns=['icustay_id', 'new_aki'])
```
